```python
import numpy as np
from scipy.optimize import fsolve

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-2, 0.1, 0.1)
```

```python
y=x**2
```

```python
a0 = 8/3

n=1
gar_1 = a0-2*(((4*n**2*np.pi**2-2)*np.sin(-2*np.pi*n)+4*np.pi*n*np.cos(-2*np.pi*n))/(np.pi**3*n**3))*np.cos(n*x)

n=2
gar_2 = a0-2*(((4*n**2*np.pi**2-2)*np.sin(-2*np.pi*n)+4*np.pi*n*np.cos(-2*np.pi*n))/(np.pi**3*n**3))*np.cos(n*x)

n=3
gar_3 = a0-2*(((4*n**2*np.pi**2-2)*np.sin(-2*np.pi*n)+4*np.pi*n*np.cos(-2*np.pi*n))/(np.pi**3*n**3))*np.cos(n*x)

n=4
gar_4 = a0-2*(((4*n**2*np.pi**2-2)*np.sin(-2*np.pi*n)+4*np.pi*n*np.cos(-2*np.pi*n))/(np.pi**3*n**3))*np.cos(n*x)

summ = (gar_1 - a0) + (gar_2 - a0) + (gar_3 - a0) + gar_4
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 16, 16
```

```python
title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "#808080",
    "family": "serif",
}

label_font = {
    "fontsize": 9,
    "family": "serif",
}

legend_font = {
    "size": 10,
    "family": "serif",
}

plt.plot(x, y, label="x^2", color="g")
plt.hlines(a0, np.min(x), np.max(x), color = 'r', label="Гармоника 0")
plt.plot(x, gar_1, label="Гармоника 1", color="r", linestyle="dashed")
plt.plot(x, gar_2, label="Гармоника 2", color="y", linestyle="dashed")
plt.plot(x, gar_3, label="Гармоника 3", color="b", linestyle="dashed")
plt.plot(x, gar_4, label="Гармоника 4", color="black", linestyle="dashed")
plt.plot(x, summ, label="Общий", color="black")


plt.title("График", fontdict=title_font)
plt.xlabel("x", fontdict=label_font)
plt.ylabel("y", fontdict=label_font)

legend = plt.legend(loc="lower right", prop=legend_font)
plt.setp(legend.get_texts(), color="DarkBlue")

plt.grid(color="lightgrey")

plt.show()
```
