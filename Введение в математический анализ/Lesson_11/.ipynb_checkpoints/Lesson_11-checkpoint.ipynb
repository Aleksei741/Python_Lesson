```python
import numpy as np
from scipy.optimize import fsolve

import matplotlib.pyplot as plt
%matplotlib inline

from pylab import rcParams
rcParams['figure.figsize'] = 16, 16
```

```python
x = np.arange(-np.pi, np.pi, 0.05)
```

```python
#Функция
y = np.e**x

a0 = ((np.e**np.pi) - (np.e**-np.pi))/(2*np.pi)
#an = ((np.e**np.pi)*np.cos(n*np.pi)-(np.e**-np.pi)*np.cos(-np.pi*n))/(np.pi*(n**2+1))
#bn = (-n*(np.e**np.pi)*np.cos(n*np.pi)-n*(np.e**-np.pi)*np.cos(-np.pi*n))/(np.pi*(n**2+1))

#Фурье fx
_n_ = 5
fx = a0
for n in range(1, _n_):
    an = ((np.e**np.pi)*np.cos(n*np.pi)-(np.e**-np.pi)*np.cos(-np.pi*n))/(np.pi*(n**2+1))
    bn = (-n*(np.e**np.pi)*np.cos(n*np.pi)-n*(np.e**-np.pi)*np.cos(-np.pi*n))/(np.pi*(n**2+1))
    fx = fx + (an*np.cos(n*x) + bn*np.sin(n*x))

#Маклорена mx
_n_ = 5
mx = 1
for n in range(1, _n_):
    mx = mx + (x**n)/np.math.factorial(n)
```

```python
title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "#808080",
    "family": "serif",
}

label_font = {
    "fontsize": 9,
    "family": "serif",
}

legend_font = {
    "size": 10,
    "family": "serif",
}

plt.plot(x, y, label="e^x", color="g")
plt.plot(x, fx, label="Фурье", color="r", linestyle="dashed")
plt.plot(x, mx, label="Маклорена", color="y", linestyle="dashed")



plt.title("График", fontdict=title_font)
plt.xlabel("x", fontdict=label_font)
plt.ylabel("y", fontdict=label_font)

legend = plt.legend(loc="lower right", prop=legend_font)
plt.setp(legend.get_texts(), color="DarkBlue")

plt.grid(color="lightgrey")

plt.show()
```
