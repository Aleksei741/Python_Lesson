```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import datetime
```

```python
df = pd.read_csv('opendata.csv', index_col=False, encoding='Windows-1251')
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
parameters = df['name'].unique()
print('Выберите параметр из списка ниже:\n')
for param in parameters:
    print(param)
```

```python
parameter = 'Средняя пенсия'
```

```python
data_full_regions = df.loc[df['name'] == parameter, :]
```

```python
data_full_regions.head()
```

```python
regions = data_full_regions['region'].unique()
print('Выберите регион из списка ниже:\n')
for region in regions:
    print(region)
```

```python
region = 'Санкт-Петербург'
```

```python
data_sort = data_full_regions.loc[data_full_regions['region'] == region]
```

```python
data_sort.head()
```

```python
print(f"Выберите период между {data_sort['date'].min()} и {data_sort['date'].max()}")
```

```python
date_min = '2017-01-15'
date_max = '2018-12-15'
```

```python
result = data_sort.loc[(data_sort['date'] >= date_min) & (data_sort['date'] <= date_max), :]
result
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 14, 9

title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "#808080",
    "family": "serif",
}

label_font = {
    "fontsize": 9,
    "family": "serif",
}

legend_font = {
    "size": 10,
    "family": "serif",
}

plt.plot(result['date'], result['value'], label=parameter)

#plt.title("График", fontdict=title_font)
plt.xlabel("Дата", fontdict=label_font)
plt.ylabel("Значение", fontdict=label_font)

legend = plt.legend(loc="lower right", prop=legend_font)
plt.setp(legend.get_texts(), color="DarkBlue")
plt.xticks(rotation='vertical')
plt.grid(color="lightgrey")

plt.show()

```
